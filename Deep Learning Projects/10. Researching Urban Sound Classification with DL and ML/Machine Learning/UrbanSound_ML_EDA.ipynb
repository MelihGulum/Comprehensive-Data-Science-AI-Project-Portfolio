# **OUTLIER**

* I'm wondering which class the maximum values of numeric variables belong to.
* **I will not replace the outliers.** Because it is a still mystery among Audio Signal Domain researchers that these are whether outliers or not.

```python
for col in df.columns:
  if df[col].dtypes in ['int674', 'float64']:
    print(col, df.loc[df.index == df[col].idxmax()]['label'].values[0])
```

```python
def outlier_thresholds(dataframe, variable, low_quantile=0.05, up_quantile=0.95):
    quantile_one = dataframe[variable].quantile(low_quantile)
    quantile_three = dataframe[variable].quantile(up_quantile)
    interquantile_range = quantile_three - quantile_one
    up_limit = quantile_three + 1.5 * interquantile_range
    low_limit = quantile_one - 1.5 * interquantile_range
    return  low_limit, up_limit
```

```python
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
```

```python
for col in df.columns[2:62]:
    print(col, check_outlier(df, col))
```

# **VISUALIZATION**

```python
ordered_class = df.groupby('label').count().sort_values(by='label').index
```

```python
fig = plt.figure(figsize=(15, 7))
ax = sns.countplot(data = df, x = 'label', order = ordered_class, palette=sns.color_palette('tab20'))

for i in ax.patches:
    ax.text(x = i.get_x() + i.get_width()/2, y = i.get_height()/2,
            s = int(i.get_height()),
            ha = 'center', va='center', size = 14, weight = 'bold', rotation = 0, color = 'white',
            bbox=dict(boxstyle="round,pad=0.5", fc='pink', ec="pink", lw=2))

plt.title('Number of Samples of Each Class',fontsize=16)
plt.show()
```

```python
df_temp = df.groupby('label')['lenght'].mean().reset_index()
fig = plt.figure(figsize=(15, 7))

ax = sns.barplot(x=df_temp['label'], y=df_temp['lenght'],
                 data = df, order = ordered_class, palette=sns.color_palette('tab20'))

for i in ax.patches:
    ax.text(x = i.get_x() + i.get_width()/2, y = i.get_height()/2,
            s = f"{round(i.get_height(),2)}",
            ha = 'center', va='center', size = 14, weight = 'bold', rotation = 0, color = 'white',
            bbox=dict(boxstyle="round,pad=0.5", fc='pink', ec="pink", lw=2))

plt.title('Average Duration of Per Class',fontsize=16)
plt.show()
```

```python
df_temp = df.groupby('label')['filename'].count().reset_index()

plt.figure(figsize=(20, 8))

patches, texts, autotexts = plt.pie(df_temp['filename'],
                                    wedgeprops = {'linewidth' : 1,
                                                  'edgecolor' : 'white' },
                                    autopct = '%.2f%%',
                                    labeldistance = 1.07,
                                    counterclock=False,
                                    startangle=-270,
                                    colors=sns.color_palette('tab20'))
plt.legend(df_temp['label'], loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), borderpad=1, fontsize=14)

plt.setp(autotexts, **{'color':'white', 'weight':'bold', 'fontsize':12})
plt.title('Ratio of Each Class',fontsize=16)
plt.tight_layout()
plt.show()
```
